```python
import csv
import operator
import copy
from token import EQUAL
from builtins import sorted
```

```python
#open the csv file
reader = csv.reader(open(r'C:\Users\phili\Desktop\Data\Largest_Cities_CSV.csv'), delimiter=",")

#Initialize the variable
u = 'State - Place'
x = 'Year'
y = 'Population'
z = 'Denny - Rank'
rows_so_far = 0
c = 0

#Initializing the 2D list
pool = []
pool.append([])
```

```python
#loop through the csv file and append the list
for row in reader:
    if rows_so_far==0:
        rows_so_far +=1
        header = row
        
#populate header
#header for the csv file should include
#state-city, year, population, rank
        for j in range(0,4):
            if j==0:
                pool.append([])
                pool[0].append(u)
            if j==1:
                pool[0].append(x)
            if j==2:
                pool[0].append(y)
            if j==3:
                pool[0].append(z)
    else:
#populate items
#append the population a city for each year
        for i in range(len(row)-2):
            a=len(pool)
            if not row==[]:
                if i==0 or i>=1:
                    item=copy.deepcopy(row)
                    r=copy.deepcopy(row)
                    
#populate the values for a column
                    for j in range(0,4):
                        if item[i+2] != '':
                            if j==0:
                        
#concatenate the state and city value
                                r[0]=item[j+1]+' - '+item[j]
                                pool.append([])
                                pool[a-1].append(r[0])
            
            
#populate year value
                            if j==1:
                                pool[a-1].append(int(header[i+2]))
            
#popluate population value
#Incase population is nil, do not append the record
                            if j==2:
                                if item[i+2]=='':
                                    pool[a-1].append(int(0))
                                else:
                                    pool[a-1].append(int(item[i+2]))
#Initialise the rank to zero                                
                            if j==3:
                                pool[a-1].append(int(0))
            
    rows_so_far+=1
```

```python
#check the lenght of the list
a = len(pool)

#populate all the list with all values except header as
#Sorting cannot be done on integer values when one record has character value 
list = pool[1:a-1]

#sort the year by year & population
list.sort(key = lambda b: (b[1],b[2]), reverse=True)

#add header to the list
list1= []
list1.append([])
list1[0] = pool[0]
list1[1:a-1] = list[0:a-2]

#convert list into csv file
mycsv = csv.writer(open(r'C:\Users\phili\Desktop\Data\Denny.csv','w', newline=''))
for row in list1:
#fetching the row index
#no calculation is done at header level
    e=list1.index(row)
#compare the year value of the current record  with the year value of the previous record
#if same calculate the rank, if different assigns the rank as one
    if row[1] != c and e !=0:
        v=1
        c=row[1]
        row[3]=v
    else:
        if row[1] == c and e !=0:
            v+=1
            row[3]=v
#write row to csv
    mycsv.writerow(row)
```
